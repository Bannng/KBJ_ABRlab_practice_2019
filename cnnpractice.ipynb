```
from tensorflow.keras.datasets.cifar10 import load_data
```

```
(x_train, y_train), (x_test, y_test) = load_data()
```

```
import tensorflow as tf
import numpy as np

X = tf.placeholder(tf.float32, [None, 32,32,3])
Y = tf.placeholder(tf.float32, [None,10])

conv1 = tf.layers.conv2d(X, 64, (3,3), padding="same",activation = None, use_bias = True)
conv_act1 = tf.nn.relu(conv1)
pool1 = tf.layers.max_pooling2d(conv_act1, (2,2), strides=(2,2))
print(pool1.shape)

conv2 = tf.layers.conv2d(pool1, 64, (3,3), padding="same", activation = None, use_bias = True)
conv2_act2 = tf.nn.relu(conv2)
pool2 = tf.layers.max_pooling2d(conv2_act2, (2,2), strides=(2,2))
print(pool2.shape)

conv3 = tf.layers.conv2d(pool2, 128, (3,3), padding="same", activation = None, use_bias=True)
conv3_act3 = tf.nn.relu(conv3)
pool3 = tf.layers.max_pooling2d(conv3_act3, (2,2), strides=(2,2))
print(pool3.shape)

conv4 = tf.layers.conv2d(pool3, 64, (3,3), padding="same", activation=None, use_bias = True)
conv4_act4 = tf.nn.relu(conv4)
print(conv4_act4.shape)

flat = tf.layers.flatten(conv4_act4)
print(flat.shape)

fc1 = tf.layers.dense(flat, 128, use_bias=True)
fc1_act1 = tf.nn.relu(fc1)
print(fc1_act1.shape)

pred = tf.layers.dense(fc1_act1, 10 , use_bias=True)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred, labels=Y))
optimizer = tf.train.AdamOptimizer(0.0001).minimize(cost)
```

```
init = tf.global_variables_initializer()
with tf.Session() as sess:
  sess.run(init)
  
  num_epoch = 20
  batch_size = 2500
  total_batch = int(len(x_train)/batch_size)
  
  for epoch in range(num_epoch):
    total_cost = 0
    
    for i in range(total_batch):
      if i == (total_batch-1):
        order = list(range(len(x_train)))
        import random
        random.shuffle(order)
        x_train = x_train[order]
        y_train = y_train[order]
        
      batch_xs = x_train[i*batch_size:(i+1)*batch_size].reshape(batch_size,32,32,3) 
      batch_ys = np.eye(10)[y_train[i*batch_size:(i+1)*batch_size]].reshape(batch_size,10)
      #print(batch_ys.shape)
      
      _, cost_val = sess.run([optimizer, cost], feed_dict={X: batch_xs, Y: batch_ys})
      total_cost += cost_val
      
    print('Epoch:', '%03d' % (epoch+1), 'Avg. cost=','{:.8f}'.format(total_cost / total_batch))
  
  print('최적화 완료!')  
  
  is_correct = tf.equal(tf.argmax(pred,1), tf.argmax(Y,1))
  accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
  print('정확도:',sess.run(accuracy,feed_dict={X: x_test.reshape(len(x_test),32,32,3), Y: np.eye(10)[y_test].reshape(len(y_test),10)}))
  

```
